<b><span style="color:green">Answer:</span></b>
    
Below are the functions for f(x) along with its first and second derivative functions.    
$$f(x) = x^3−12x^2−6$$
$$f'(x) = 3x^2-24x$$
$$f''(x) = 6x-24$$

Eyeballing the graphs below, it looks as though we have a local maximum at the point $(0, -6)$. 

The FOC for optimality tells us that we have the minimum/maxiumum point when $f'(x=x*) = 0$. Note that $0$ is the value of $x$ for which $f'(x) = 0$, also known as the root. We can thus use $f(x)$ to calculate that $f(0) = 0-0-6 = -6$, giving us the point $(0, -6)$ identified graphically.

The second order optimality conditions (SOC) are that if $f''(x=x*) < 0$, then $x*$ is a maximum. Otherwise, if $f''(x=x*) > 0$, then $x*$ is a minimum. In our graph, we see that $f''(0) < 0$, confirming our claim that we see a local maximum at $(0, -6)$.

A Hessian matrix is simply a square matrix of second-order partial derivatives. Since we only have one variable, x, the Hessian matrix is simply the partial second derivative of x.

$$H = \begin{bmatrix}\frac{\partial^2f}{\partial x_1^2}\end{bmatrix}$$
$$H = \begin{bmatrix}6x-24\end{bmatrix}$$

```python
import numpy as np
import matplotlib.pyplot as plt

# Make matplotlib appear in the workbook
%matplotlib inline
plt.figure(figsize=(10,8))

# define function
ax = plt.subplot(311)
t = np.arange(-6.0, 6.0, 0.1)
s = [(x*x*x - 12*x*x - 6) for x in t]
ax.set_title("f(x)")

# Create and annotate line
line = plt.plot(t, s, lw=2)
plt.axhline(y=0.,color='k',ls='dashed')
plt.annotate('local max', xy=(0, -6), xytext=(1, -250),
            arrowprops=dict(facecolor='black', shrink=0.05),)

# First derivative (3x^2 + 24x)
ay = plt.subplot(312)
u = [(3*x*x - 12*2*x) for x in t]
line = plt.plot(t, u, lw=2)
plt.axhline(y=0.,color='k',ls='dashed')
plt.annotate('SOC root', xy=(0, -6), xytext=(1, 150),
            arrowprops=dict(facecolor='black', shrink=0.05),)
ay.set_title("f '(x)")

# First derivative (3x^2 + 24x)
az = plt.subplot(313)
v = [(3*2*x - 12*2) for x in t]
line = plt.plot(t, v, lw=2)
plt.axhline(y=0.,color='k',ls='dashed')
plt.annotate('negative value', xy=(0, -25), xytext=(1, -40),
            arrowprops=dict(facecolor='black', shrink=0.05),)
az.set_title("f ''(x)")

plt.show()
```

<b><span style="color:green">Answer:</span></b>
    
$$f(x) = x^3 + 2x -4$$
$$f'(x) = 3x^2 + 2$$

To do the Newton-Raphson method, we need a point on the curve and the tangent slope. We can plug in to get the point, and calculate the tangent slope as the derivative. 
$$f(x) = x^3 + 2x -4$$
$$f(1) = 1^3 + 2 - 4 = -1$$

The point is $(1, -1)$

$$f'(x) = 3x^2 + 2$$
$$f'(1) = 3 + 2 = 5$$

The slope is $5$

With the point and the slope, we can calculate the line via $y = mx + b$
$$-1 = 5*1 + b$$
$$-6 = b$$
$$y = 5x - 6$$

We then solve for $y = 0$ to get $xt2$.
$$0 = 5x - 6$$
$$6 = 5x$$
$$6/5 = x = 1.2$$

This matches the prompt, telling us the solution would be 1.2.

<b><span style="color:green">Answer:</span></b>

An optimization problem is convex if the objective function being optimized is a convex function. A object is convex if, for every pair of points within the object, all of the points on a straight line connecting the pair is also in the object. Similarly, a convex function is defined as one where, if you take a pair of points on the function, the points on a line segment connecting the pair will all fall above the function itself.

The first order condition for convex is that the function is globally above the tangent at y. The second order optimality conditions are not important. The first order condition is enough alone to know that the minimum found is a global minimum, not a local minimum or a maximum. This is extremely convienent to machine learning, because we do not need to test the second order equation at all to know we have found a global minimum after gradient descent. It is also useful because the sum of any convex functions will also be convex. Hence, when we have a convex cost function and a convex regularization term, the sum of the two will also be convex.
